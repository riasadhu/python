### Step 3. Assign it to a variable called users.

```python
users=pd.read_csv('occupation.csv', sep='|')
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
import numpy as np
first=users.groupby(['occupation','gender']).count().reset_index().groupby(['occupation'])['user_id'].sum()
first_1=users.groupby(['occupation','gender']).count().reset_index()
total=first.reset_index()
requiredtable=pd.merge(total,first_1,how='inner',on='occupation')
requiredtable=requiredtable.assign(ratio=requiredtable.user_id_y/requiredtable.user_id_x)[['occupation','gender','ratio']]
requiredtable.where(requiredtable.gender=='M').dropna().sort_values(by='ratio')
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.pivot_table(index = 'occupation', values = 'age', columns='gender'
                   ,aggfunc = 'mean')
```

### Step 8.  For each occupation present the percentage of women and men

```python
first=users.groupby(['occupation','gender']).count().reset_index().groupby(['occupation'])['user_id'].sum()
first_1=users.groupby(['occupation','gender']).count().reset_index()
total=first.reset_index()
requiredtable=pd.merge(total,first_1,how='inner',on='occupation')
requiredtable=requiredtable.assign(ratio=(requiredtable.user_id_y/requiredtable.user_id_x)*100)[['occupation','gender','ratio']]
```

```python
requiredtable.pivot(index='occupation',columns='gender',values='ratio').rename(columns={'F':'F_%','M':'M_%'})
```
